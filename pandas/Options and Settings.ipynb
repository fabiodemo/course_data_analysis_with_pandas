# Options and Settings

```python
import pandas as pd 
import numpy as np
```

## Changing Options with Attributes and Dot Syntax

rand integers from 0 to 1000, and the array size we want to return \
 (from, to, [rows, columns]) 

```python
data = np.random.randint(0, 100, [1000, 50])
df = pd.DataFrame(data)
df.tail(3)
```

Setting how many rows will be shown in the DataFrame

```python
pd.options.display.max_rows = 18
```

```python
df
```

```python
pd.options.display.max_rows = 5
```

```python
df
```

```python
pd.options.display.max_columns
```

```python
pd.options.display.max_columns = 4
```

## Changing pandas Options with Methods

```python
data = np.random.randint(0, 100, [1000, 50])
df = pd.DataFrame(data)
df.tail(3)
```

```python
pd.get_option("max_rows")
```

```python
pd.get_option("max_columns")
```

```python
pd.set_option("max_rows", 16)
```

```python
df
```

```python
pd.set_option("max_columns", 16)
```

```python
df
```

```python
pd.options.display.max_columns
```

```python
pd.options.display.max_columns = 10
```

```python
df
```

```python
pd.reset_option("max_columns")
```

```python
pd.reset_option("max_rows")
```

```python
df
```

```python
pd.describe_option("max_columns")
```

## The precision Option

Doesn't modify the number, just hides the complexity of the number by framing a simpler floating number

```python
df = pd.DataFrame(np.random.randn(5, 5))
df
```

Get the precision of our float number

```python
pd.get_option("precision")
```

Changing the precision to 2 digits after decimal point in an entire dataframe

```python
pd.set_option("precision", 2)
```

```python
df
```
